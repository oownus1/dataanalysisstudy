```python
import pandas as pd
```

```python
import json
```

```python
file_name = "access_log_20220602.txt"

file = open(file_name, "r")

data = []

order = ["A","CHECK","C"] #CHECK라고 지정한 컬럼부븐에서 접속 정보를 확인하였습니다


# for문으로 로그데이터를 띄어쓰기로 나누고 이를 컬럼과 매핑시켜 데이터를 딕셔너리고 가져왔습니다
for line in file.readlines():
    details = line.split("(")
    details = [x.strip() for x in details]
    structure = {key:value for key, value in zip(order, details)}
    data.append(structure)
    
# 딕셔너리 형태로 데이터를 불러와주었습니다. 
# json.dumps로 형태를 맞춰주었습니다.
for entry in data:
    print(json.dumps(entry, indent = 20))
```

```python
#frame2라는 이름으로 데이터 프레임을 만들었습니다.
frame2= pd.DataFrame(data)
```

```python
# 전체 데이터는 6612개 중 6200개의 접속 정보를 볼 수 있었습니다. -> 6612 개중에서 6200을 뺀 412개의 결측치가 있는 것을 알 수 있었습니다.
print(frame2.info())
```

```python
# 데이터프레임 출력
frame2
```

```python
# 시각화를 위한 라이브러리인 matplotlib과 seaborn을 불러왔습니다.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#접속정보 데이터들을 막대그래프로 시각화해보았습니다
#접속정보가 많아서 시각화로 보기 힘든 부분이었습니다.
sns.countplot(x='CHECK', data=frame2)
```

```python
#각 접속 정보 형태와 개수를 다음과 같이 출력해주었습니다
# 접속 정보 형태와 개수들이 출력된 것을 보실 수 있으십니다.
frame2.groupby('CHECK').size()
```

```python
# 크롬으로 접속한 접속정보 데이터프레임을 추출해서 chrome_count라는 변수를 지정하여 출력해주었습니다
# 크롬 접속정보에 들어가는 단어인 AppleWebKit으로 찾아주었습니다
chrome_count = frame2[frame2['CHECK'].str.contains('AppleWebKit', na = False)]
chrome_count
```

```python
alist = chrome_count['A'].str.split(' ')
alist.head()
```

```python
chrome_count['ip'] = alist.str.get(0)
chrome_count.head()
```

```python
# 크롬이 접속한 ip건수 결과를 그래프로 출력하였습니다
sns.countplot(x='ip', data=chrome_count)
```

```python
chrome_count.groupby('ip').size()
```

```python
# 크롬접속은 6200개중 14개가 출력된 것을 보실 수 있으십니다.
chrome_count.count()
```

```python
# 익스플로어로 접속한 접속정보 데이터프레임을 추출해서 explorer_count라는 변수를 지정하여 출력해주었습니다
# 익스플로어 접속정보에 들어가는 단어인 MSIE으로 찾아주었습니다
explorer_count = frame2[frame2['CHECK'].str.contains('MSIE', na = False)]
explorer_count
```

```python
#익스플로어 접속은 6200개 중 6153개로 접속된 것을 알 수 있었습니다. 
explorer_count.count()
```

```python
alist1 = explorer_count['A'].str.split(' ')
alist1.head()
```

```python
explorer_count['ip1'] = alist1.str.get(0)
explorer_count.head()
```

```python
sns.countplot(x='ip1', data=explorer_count)
```

```python
#익스플로어 접속한 ip종류와 개수를 구하여 출력하였습니다.
result = explorer_count.groupby('ip1').size()
result
```

```python
explorer_count.groupby(explorer_count['ip1']).sum().plot.bar()
# 막대그래프 시각화가 진행되지 않았습니다. 
```
