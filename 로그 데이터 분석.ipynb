```python
import pandas as pd
```

```python
import json
```

```python
#로그 데이터를 텍스트 파일로 불러왔습니다. 변수는 file_name으로 지정하여 불러온 부분입니다
file_name = "access_log_20220602.txt"

#해당 파일을 읽기 형태인 r로 하여 읽어온 부분입니다.
file = open(file_name, "r")

# 불러온 데이터를 담을 data부분을 빈 리스트로 가져왔습니다. 이를 사용하여 매핑하여 딕셔너리 형태로 결과를 출력할 계획입니다
# 딕셔너리 형태로 결과를 출력할려는 이유는 이형태로 가져와야 데이터 프레임형태를 만들 수 있기 때문입니다. 
# 데이터 프레임 형태로 만들려는 이유는 시각화를 진행하기 위해서입니다.
data = []

# 로그데이터를 나누어서 각각의 컬럼을 만들어 주었습니다. 
# 로그데이터는 아래의 띄어쓰기를 기준으로 나눌 계획입니다
# 로그데이터 형식이 다양하게 이루어져있어서 띄어쓰기로 나누었습니다.
order = ["ip","-","-1","date","date1","POST/GET","url","HTTP","port1","port2","data"] #message


# for문으로 로그데이터를 띄어쓰기로 나누고 이를 컬럼과 매핑시켜 데이터를 딕셔너리고 가져왔습니다
for line in file.readlines():
    details = line.split(" ")
    details = [x.strip() for x in details]
    structure = {key:value for key, value in zip(order, details)}
    data.append(structure)
    
# 딕셔너리 형태로 데이터를 불러와주었습니다. 
# json.dumps로 형태를 맞춰주었습니다.
for entry in data:
    print(json.dumps(entry, indent = 20))
    
    
# 그랬을 때 아래와 같이 데이터를 불러와 줍니다. 
# 이를 .DataFrame을 사용하여 데이터 프레임형태로 결과를 출력하겠습니다. 


# 이 딕셔너리 형태로 데이터를 불러오기 위해서 다양한 방식들을 사용하였습니다
# 시행착오
# 파일을 인코딩하여 불러오는 과정에서 encoding오류가 나기도 하였고, 
# 무엇보다 json형태로 파일의 데이터를 불러오려고 했을 때 발생하는 JSONDecodeError이 어려웠습니다.
# JSONDecodeError: Expecting value: line 1 column 2 (char 1)
# 이부분을 위와 같이 불러옴으로써 해결하였습니다. 
# 이과정을 통해 HTTP통신, 로그데이터 문제인지 다양한 문제들을 파악해보는 과정을 겪었습니다. 
```

```python
#위에서 json형태로 불러온 파일을 .DataFrame으로 데이터프레임을 만들어 주어서 frame1이라는 변수에 저장하였습니다
frame1 = pd.DataFrame(data)
```
