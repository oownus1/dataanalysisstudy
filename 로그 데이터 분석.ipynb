```python
import pandas as pd
```

```python
import json
```

```python
#    .  file_name   
file_name = "access_log_20220602.txt"

#    r   .
file = open(file_name, "r")

#    data   .        
#             . 
#        .
data = []

#      . 
#      
#      .
order = ["ip","-","-1","date","date1","POST/GET","url","HTTP","port1","port2","data"] #message


# for         
for line in file.readlines():
    details = line.split(" ")
    details = [x.strip() for x in details]
    structure = {key:value for key, value in zip(order, details)}
    data.append(structure)
    
#    . 
# json.dumps  .
for entry in data:
    print(json.dumps(entry, indent = 20))
    
    
#       . 
#  .DataFrame     . 


#         
# 
#     encoding  , 
#  json       JSONDecodeError .
# JSONDecodeError: Expecting value: line 1 column 2 (char 1)
#     . 
#   HTTP,       . 
```

```python
# json   .DataFrame    frame1  
frame1 = pd.DataFrame(data)
```
