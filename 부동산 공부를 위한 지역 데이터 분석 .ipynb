```python
#박선우 부동산 공부를 조금씩 시작해보기 위해서 행정구역 정보를 읽고 분석하는 과정을 진행해았습니다. 
#관련 정보가 담긴 데이터를 가져와서 간단한 분석과 시각화를 진행하며 공부를 행정구역에 대한 공부를 해보았습니다

#필요 라이브러리를 불러와서 데이터를 로드
#데이터에서 부동산공부에 활용되는 면적, 행정구역과 광역시도 명칭, 인구수 등을 볼 수 있는 데이터입니다
#이를 가지고 분석을 진행해보겠습니다 ㅎㅎ
import pandas as pd  

df = pd.read_csv('./data_draw_korea.csv')
df
```

```python
# 데이터의 통계값 보기
df.describe()
```

```python
#우리나라이 행정구역 전체를 봐보았습니다 (중복제거)
#못둘어본 행정구역이 정말 많은 것을 알게 되었습니다....
#이중에 부동산 기회는 어디에 있을까요ㅎㅎ
df['행정구역'].unique()
```

```python
#광역시도 출력해보았습니다. 
df['광역시도'].unique()
```

```python
#지역 출력
# 못들어본 지역이 엄청 많은 것을 알 수 있었습니다.. 어느곳을 임장을 가봐야 잘 가보았다고 소문이 날까요.. ㅎㅎ
df['shortName'].unique()
```

```python
#제가 살고있는 광역시도가 서울특별시인 행과 열을 모두 출력해보았습니다
#인구수가 높은순으로 ascending을 지정하여 출력
df.loc[df['광역시도'] == '서울특별시'].sort_values(by='인구수', ascending=False).reset_index(drop=True)
# 송파구 인구수가 가장많은 것을 데이터를 통해 알게 되었습니다. 
# 현재 제가 거주하는 영등포구가 14위네요 이렇게 데이터를 알 수 있어서 너무 좋습니다
```

```python
#경기도에 속한 행정구역의 행정구역, 인구수, 면적을 봐보고 싶어서 출력해보았습니다
df.loc[df['광역시도'] == '경기도', ['행정구역', '인구수', '면적']].sort_values(by='인구수', ascending=False).reset_index(drop=True)
#경기도의 인구수 순으로 출력을 해보았습니다. 수원시 인구수가 많은 것을 알 수 있었습니다 
#결과에 나온 곳들의 지역을 잘 선정하여 임장을 다녀와봐야겠습니다.
```

```python
#인구수가 전국 평균보다 작은 지역을 선택하고 결과출력해보겠습니다. 
#전국 평균 인구수 구하기
people_mean = df['인구수'].mean()
print(people_mean)
#평균인구수가 약 21만으로 출력되었습니다.
```

```python
#전체 인구수 합도 구해보았습니다
# 5500만정도로 알았는데 이 데이터의 수치는 다르게 나타나네요 ㅎㅎ
df['인구수'].sum()
```

```python
#평균 인구수보다 낮은 지역 
people_mean_under = df.loc[df['인구수'] <= people_mean, ['인구수','행정구역','면적','광역시도']]
```

```python
#평균 인구수보다 낮은 행정구역을 알 수 있었습니다. 
people_mean_under
```

```python
#인구수와 면적간에 관련성이 있는지 살펴보기 위해서 상관계수를 구헤보았습니다
#1에 가까울수록 연관성이 크고 -1에 가까울수록 연관성이 작습니다
df['면적'].corr(df['인구수'])
#큰 상관관계가 있는 것은 아닌것으로 출력됩니다.
```

```python
#경기도 면적과 인구수의 상관계수 구해보기
gyunggi_df = df.loc[df['광역시도'] == '경기도']
gyunggi_df['면적'].corr(gyunggi_df['인구수'])
```

```python
#경기도 면적과 인구수의 상관계수 구해보기
seoul_df = df.loc[df['광역시도'] == '서울특별시']
seoul_df['면적'].corr(seoul_df['인구수'])
#경기도와 비교했을 때 서울특별시에서는 인구수와의 관계 관련성이 있는 것을 알 수 있었습니다
```

```python
#광역시도별 인구수 합계 구해보기
df['인구수'].groupby(df['광역시도']).sum().sort_values(ascending=False)
#경기도의 인구수가 가장 높은 결과를 알 수 있었다. 
#시리즈 형태
```

```python
#데이터 프레임 형태로 광역시도별 인구수의 합계 구해보기
df.groupby('광역시도').sum()
```
